## Visualize QuickDraw model results

```python
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
import os, json
import pandas as pd
```

```python
def load_json(path, filetype):
    f = open(os.path.join(path, 'model_results.json'))
    df = pd.read_json(f)
    df['type'] = filetype
    f.close()
    f = open(os.path.join(path, 'results.json'))
    results = json.load(f)
    test_loss = results['mean_outer_loss']
    test_acc = results['accuracies_after']
    f.close()
    return df, test_loss, test_acc
```

```python
data_20_5_1, test_20_loss_5_1, test_20_acc_5_1 = load_json('../models/2020-09-24_220708', "5-way, 1-shot")
data_20_5_5, test_20_loss_5_5, test_20_acc_5_5 = load_json('../models/2020-09-24_182934', "5-way, 5-shot")
data_20_5_10, test_20_loss_5_10, test_20_acc_5_10 = load_json('../models/2020-09-24_220801', "5-way, 10-shot")
```

```python
plt.figure(figsize=(15,10))
plt.subplot(2, 2, 1)
plt.plot(data_20_5_1['epoch'], data_20_5_1['train_loss'], label="5-way, 1-shot", color='xkcd:lilac')
plt.plot(data_20_5_1['epoch'], data_20_5_5['train_loss'], '-.', label="5-way, 5-shot", color='xkcd:purplish blue')
plt.plot(data_20_5_1['epoch'], data_20_5_10['train_loss'], '--', label="5-way, 10-shot", color='xkcd:vibrant purple')
plt.legend()
plt.xlabel('epochs')
plt.ylabel('loss')
plt.title("training loss")
plt.yticks(np.arange(0.4, 1.4, step=0.2))

plt.subplot(2, 2, 2)
plt.plot(data_20_5_1['epoch'], data_20_5_1['val_loss'], label="5-way, 1-shot", color='xkcd:lilac')
plt.plot(data_20_5_1['epoch'], data_20_5_5['val_loss'], '-.', label="5-way, 5-shot", color='xkcd:purplish blue')
plt.plot(data_20_5_1['epoch'], data_20_5_10['val_loss'], '--', label="5-way, 10-shot", color='xkcd:vibrant purple')
plt.plot(50, test_20_loss_5_1, 'o', label="test 5-way, 1-shot", color='xkcd:lilac')
plt.plot(50, test_20_loss_5_5, 'o', label="test 5-way, 5-shot", color='xkcd:purplish blue')
plt.plot(50, test_20_loss_5_10, 'o', label="test 5-way, 10-shot", color='xkcd:vibrant purple')

plt.legend()
plt.xlabel('epochs')
plt.ylabel('loss')
plt.title("validation loss")
plt.yticks(np.arange(0.4, 1.4, step=0.2))

plt.subplot(2, 2, 3)
plt.plot(data_20_5_1['epoch'], data_20_5_1['train_acc'], label="5-way, 1-shot", color='xkcd:lilac')
plt.plot(data_20_5_1['epoch'], data_20_5_5['train_acc'], '-.', label="5-way, 5-shot", color='xkcd:purplish blue')
plt.plot(data_20_5_1['epoch'], data_20_5_10['train_acc'], '--', label="5-way, 10-shot", color='xkcd:vibrant purple')
plt.legend()
plt.xlabel('epochs')
plt.ylabel('accuracy')
plt.title("training accuracy")
plt.yticks(np.arange(0.4, 1.1, step=0.1))

plt.subplot(2, 2, 4)
plt.plot(data_20_5_1['epoch'], data_20_5_1['val_acc'], label="5-way, 1-shot", color='xkcd:lilac')
plt.plot(data_20_5_1['epoch'], data_20_5_5['val_acc'], '-.', label="5-way, 5-shot", color='xkcd:purplish blue')
plt.plot(data_20_5_1['epoch'], data_20_5_10['val_acc'], '--', label="5-way, 10-shot", color='xkcd:vibrant purple')
plt.plot(50, test_20_acc_5_1, 'o', label="test 5-way, 1-shot", color='xkcd:lilac')
plt.plot(50, test_20_acc_5_5, 'o', label="test 5-way, 5-shot", color='xkcd:purplish blue')
plt.plot(50, test_20_acc_5_10, 'o', label="test 5-way, 10-shot", color='xkcd:vibrant purple')

plt.legend()
plt.xlabel('epochs')
plt.ylabel('accuracy')
plt.yticks(np.arange(0.4, 1.1, step=0.1))
plt.title("validation accuracy")
```
